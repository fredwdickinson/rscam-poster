An attempt to replicate the images in Bengt Fornberg's [seminar slides](https://www.newton.ac.uk/seminar/40010/), part of a series in 'Complex analysis: techniques, applications and computations - perspectives in 2023' by the Newton Institute. 

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as patches
```

```python
# Figsize to be adjusted at end
fig, ax = plt.subplots()

# Change from green only for the first real-line case
ss_colour = "green"

# ss 'Stencil Size', g for 'Green'
ss = 3
xg, yg = np.meshgrid(range(ss), range(ss))
xg, yg = xg.flatten(), yg.flatten()

# Number of red points.
line_total = 7
xr = np.arange(1 + (ss //2), 1 + (ss // 2) + line_total)
yr = np.ones(line_total) * (2 + (ss // 2))

# Bounding box from 0-0, want grid to be 2 above.
boundary = patches.Rectangle(
    # Bottom-left corner/
    (0, 0),         
    # With and height.
    # Width: stencils + line + 2 for border (either side).
    # Height: 2 border (top and bottom) + stencil size.
    (2*ss - 1 + (line_total - 2*(ss // 2))), (1 + ss),          
    linewidth = 1,
    edgecolor = 'black',
    facecolor = 'none',
    linestyle = '--' 
)

# Shift the green xs by 1 to have padding with.
xg = xg + 1

# Green stencils.
ax.scatter(xg, yg + 1, c = ss_colour, s = 40)
ax.scatter(xg + (ss // 2) + line_total - 2 ,  yg + 1, c = ss_colour, s = 40, label = 'weight $1/2$')
ax.scatter(xr, yr - 1, c = 'red', s = 40, label = 'weight $1$')

ax.add_patch(boundary)

ax.set_aspect('equal')
ax.axis('off')

save = False 
if save:
    plt.savefig("traps2a.pdf", format = "pdf", bbox_inches='tight', pad_inches = 0)

plt.show()
```
